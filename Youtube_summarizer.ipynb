```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
def get_video_id(url_link):
  return url_link.split("watch?v=")[-1]
```

```python
Video_id = get_video_id("https://www.youtube.com/watch?v=KDorKy-13ak")
#THIS LINK WHAT WORKS IS ONLY THE LINK IN ABOVE , LINK PROVIDED BY YOUTUBE AFTER CLICKING SHARE WONT WORK FOR SOME REASON
```

```python
transcripts = YouTubeTranscriptApi.get_transcript(Video_id)
#this thingy sometimes wont work because youtube only stupid and blocks certain ips
#you can either use proxy but too time consuming or use vpn which u dont have
#WHAT I DID WAS CHANGE RUNTIME TILL IT DYNAMICALLY ALLOCATES IP WHICH WORKS
```

```python
transcripts_joined = " ".join([line['text']for line in transcripts])
```

```python
from rpunct import RestorePuncts

rpunct = RestorePuncts()  # Force CPU usage
```

```python
prompt = f'"Answer questions based on: \ntext = "{transcripts_joined}. "tryna strike a cord and it\'s probably a" what comes after this ?"'
```
